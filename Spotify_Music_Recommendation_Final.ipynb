# Steps To Make Recommendation System

- obtain numerical values from song dataset
- scale the values using standard scaler
- make function to obtain song and numerical values of song from Spotify API
- scale numerical values then find mean of values (centre of song cluster)
- calculate distances from centre and other data points in dataset
- order distances from shortest to longest

### Using Standardisation

```python
from sklearn.preprocessing import StandardScaler

def numeric_scaler_df(df):
    
    years = []

    for date in df['release_date'].values:

        if group := re.match(r'([1-9][0-9][0-9][0-9])', date):
                year = group[0]

                years.append(int(year))

    df['year'] = years

    df1 = df.drop(columns='release_date')

    df_empty = pd.DataFrame()

    df_numeric = df1.select_dtypes(np.number)

    X = df_numeric.values

    numeric_dict = df_numeric.max().to_dict()

    min_val = df_numeric.max().min()

    for key in numeric_dict.keys():
        max_val = numeric_dict[key]

        if key == 'year':
            df_empty[key] = df1[key].values
            continue

        if max_val > min_val*10:
            scaler = StandardScaler()
            X = df1[key].values.reshape(-1,1)
            X_scaled = scaler.fit_transform(X)
            df_empty[f'{key}_scaled'] = X_scaled
        else:
            df_empty[key] = df1[key].values
    
    df_object = df1.select_dtypes(np.object_)
    
    df_concat = pd.concat([df_object, df_empty], axis=1)
    
    return df_concat

def get_song_metrics(song, year, df_songs):
    # not yet implemented api song request
    
    dframe = df_songs.loc[((df_songs['name'] == song) & (df_songs['year'] == year)), :]
    dframe.drop(columns='year', inplace =True)
    
    dframe = dframe.select_dtypes(np.number)
    
    song_val = dframe.iloc[0, :].values
    
    if len(song_val) == 0:
        raise Exception("No song with that name and/or year")
    else:
        
        return song_val.reshape(1,-1)

def song_centre(song_val):
    
    song_vectors = []
    
    song_vectors.append(song_val)
    
    song_matrix = np.array(list(song_vectors))

    return np.array([np.mean(song_matrix, axis=0)]).reshape(1,-1)

def distance_df(centre, X_songs, n = 10):

    distances = cdist(centre, X_songs, 'euclidean')

    index = list(np.argsort(distances)[:, 1:n+1][0])

    return index


def recommend(song, year, df_songs, n):
    
    # pass df to scale
    df_scaled = numeric_scaler_df(df_songs)
    
    # get song metrics
    song = get_song_metrics(song, year, df_scaled)
    
    # pass song metrics into find song centre to find centre
    centre = song_centre(song)
    
    # obtain all song metrics
    X_songs = df_scaled.select_dtypes(np.number).drop(columns='year').values
    
    # determine distances and return dataframe with n rows of recommendations
    index = distance_df(centre, X_songs, n)
    
    return df_songs.iloc[index]
#     return df_songs.iloc[index]['name'].values # use to return names of songs only
```
